## Data exploration and actions

```python
#Data Encoding
data['Sex'] = data['Sex'].replace({'F':0, 'M':1})
data['ExerciseAngina'] = data['ExerciseAngina'].replace({'N': 0, 'Y': 1})
```

```python
data.head()
```

```python
data2 = pd.get_dummies(data, drop_first=True)
for i in data2.columns:
    if data2[i].dtype == 'uint8':
        data2[i] = data2[i].astype('int64')
```

```python
data2.head()
```

```python
data2.info()
```

```python
fig = plt.subplots(figsize=(15,10))
sns.heatmap(data2.corr(), annot=True, linewidths=1.5)
```

<h4>All features are important and no data is missing so, nothing to dropped out.</h4>

```python
data2.describe().T
```

```python
#Train Test data splitting
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.pipeline import Pipeline
y_col = 'HeartDisease'
X = data2.drop(columns=y_col)
Y = data2[y_col]

sss = StratifiedShuffleSplit(n_splits=1, test_size=.3, random_state=10)
train_indxs, test_indxs = next(sss.split(X,Y))
X_train, Y_train, X_test, Y_test = X.loc[train_indxs,], Y.loc[train_indxs,], X.loc[test_indxs,],  Y.loc[test_indxs,]
```

```python
scaler = MinMaxScaler()
X_train1 = scaler.fit_transform(X_train)
X_test1 = scaler.transform(X_test)
```

```python
all_scores = [] #To store scores from different techniques
```

```python
from sklearn.metrics import precision_recall_fscore_support as total_score
from sklearn.metrics import accuracy_score

def add_score(y_test, y_pred, score_metrix, classifier):
    precision,recall, fscore, _ = total_score(Y_test, Y_pred, average='weighted')
    accuracy = accuracy_score(y_test, y_pred)
    score_metrix.append(pd.Series({'precision':precision, 'recall':recall, 'fscore':fscore, 'accuracy':accuracy,},name=classifier))
    return score_metrix
```

<b><font color= 'red' size='3'>All the models were fine tuned while building this notebook using various techniques like looping, Grid Search and the best estimator's hyperparameters were only used in the final models. So, I didn't keep them everywhere.</font></b>

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression(solver='liblinear',C=20).fit(X_train1, Y_train)
Y_pred = lgr.predict(X_test1)
scores = add_score(Y_test, Y_pred, [], 'LGR')
```

```python
all_scores.append(scores[0])
```

### SVM 

```python
from sklearn.svm import SVC

svc_gauss = SVC(kernel='rbf', gamma=.1, C=10).fit(X_train1, Y_train)
Y_pred = svc_gauss.predict(X_test1)
scores = add_score(Y_test, Y_pred, [], 'SVC_G')
```

```python
all_scores.append(scores[0])
```

### Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

dtc = DecisionTreeClassifier().fit(X_train, Y_train)

param_grid = {'max_depth':range(1, dtc.tree_.max_depth+1, 2),
              'max_features': range(1, len(dtc.feature_importances_)+1)}

GR = GridSearchCV(DecisionTreeClassifier(random_state=10),
                  param_grid=param_grid,
                  scoring='precision',
                  n_jobs=-1)

GR = GR.fit(X_train, Y_train)

Y_pred = GR.predict(X_test)
scores = add_score(Y_test, Y_pred, [], 'DT')
```

```python
all_scores.append(scores[0])
```

## Ensemble Method

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(oob_score=True, random_state=42, warm_start=True, n_jobs=-1)
rfc.set_params(n_estimators=400)
rfc = rfc.fit(X_train, Y_train)
Y_pred = rfc.predict(X_test)
oob_error = 1 - rfc.oob_score_
scores = add_score(Y_test, Y_pred, [], 'RFC')
```

```python
all_scores.append(scores[0])
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier(n_estimators=50, random_state=42)
gbc = gbc.fit(X_train1, Y_train)
Y_pred = gbc.predict(X_test1)
scores = add_score(Y_test, Y_pred, [], 'GBC')
```

```python
all_scores.append(scores[0])
pd.concat(all_scores, axis=1)
```

<b><font color= 'green' size='3'> Among all of the tested Models, Gradient Boosting and Random Forest performs relatively equal and highest of all. So, we can select any of them as the final model.</font></b>

<b><font color= 'green' size='3'> Let us consider Random Forest as the final Model. It consists of the 400 trees with an oob error of .14. It has a precision of .91 and while checking for a heart disease, the more the precision, the more will be good and it has a low chance of overfitting in comparison to Gradient boosting.</font></b>
